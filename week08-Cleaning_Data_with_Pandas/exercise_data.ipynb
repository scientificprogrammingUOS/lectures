# Preparing the data for the interactive exercises

```python
import pandas as pd
```

```python
pokemon = pd.read_csv("data/pokemon_no_duplicates.csv")
pokemon.head()
```

```python
(pokemon
 .groupby(["Generation", "Type 1"])
 .agg({"HP": "mean"})
 .reset_index()
 .pivot(index="Generation", columns="Type 1")
 .droplevel(0, axis="columns")
 .reset_index()
 .to_csv("data/mean_hp_per_generation_and_type.csv", index=False)
)
```

```python
pokemon_single_type = (pokemon
    .assign(Type=lambda df: df["Type 1"] + " - " + df["Type 2"])
    .drop(columns="Type 2")
)
pokemon_single_type.loc[pd.isna(pokemon_single_type["Type"]), "Type"] = pokemon_single_type["Type 1"]
pokemon_single_type = pokemon_single_type.drop(columns="Type 1")
pokemon_single_type.to_csv("data/pokemon_merged_type.csv", index=False)
pokemon_single_type.head()
```

```python
(pokemon
 .groupby("Type 1")
 .agg({"Attack": ["min", "mean", "max"]})
 .stack()
 .reset_index()
 .rename(columns={"Type 1": "Type", "level_1": "statistic"})
 .to_csv("data/pokemon_attack_statistics.csv", index=False)
)
```
